```python
#Alldf = pd.concat([Jan2016df,Feb2016df,March2016df,July2016df,Aug2016df,Sept2016df,
#                   Jan2017df,Feb2017df,March2017df,July2017df,Aug2017df,Sept2017df,
#                   Jan2018df,Feb2018df,March2018df,July2018df,Aug2018df,Sept2018df,],sort=True)

Alldf = pd.concat([Jan2017df,Feb2017df,March2017df,July2017df,Aug2017df,Sept2017df,
                   Jan2018df,Feb2018df,March2018df,July2018df,Aug2018df,Sept2018df,],sort=True)
```

```python
#Convert values to string
Alldf.gender=Alldf.gender.astype(str)

#Replace values
Alldf['gender']=Alldf['gender'].replace(['0'],"Unknown")
Alldf['gender']=Alldf['gender'].replace(['1'],"Male")
Alldf['gender']=Alldf['gender'].replace(['2'],"Female")
```
